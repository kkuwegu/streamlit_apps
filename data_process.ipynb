## Tech diagram

createa a copy of csv files for 'Technology diagram' webapp

```python
# read TechDB Excel file
path = r"C:\Users\kuke\OneDrive - empa.ch\SWEET reFuel WP3 - General\Data\reFuel_TechDatabase.xlsx"
di_tech_db = pd.read_excel(path, sheet_name=None)


# get sheet 'ConvTech'
df_tech_conv_raw = di_tech_db['ConvTech']

# setup column headers
df_tech_conv_raw.columns = df_tech_conv_raw.iloc[1]
df_tech_conv_raw = df_tech_conv_raw.drop([0, 1]).reset_index(drop=True)


# extract relevant columns
df_tech_conv_raw = df_tech_conv_raw[['ehubX Tech ID', 'Description', 'Process Type',
                             'Input Carriers', 'Main Input Carrier',
                             'Output Carriers', 'Main Output Carrier', 'Input Units - Shares', 'Output Units - Shares', 'Unit Operation'
                             ]]

# remove rows without ehubX Tech ID
df_tech_conv = df_tech_conv_raw[df_tech_conv_raw['ehubX Tech ID'].notna()].copy()

df_tech_conv = df_tech_conv[df_tech_conv['Input Carriers'].notna() | df_tech_conv['Output Carriers'].notna()]

df_tech_conv
```

```python
# export to csv

df_tech_conv.to_csv('data/tech_conv.csv', index=False)
df_tech_conv.to_excel('data/tech_conv.xlsx', index=False)
```
